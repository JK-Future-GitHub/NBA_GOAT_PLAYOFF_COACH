```python
df = pd.read_csv("greatest_playoff_coach.csv", encoding="utf-8-sig")

df = df[df['playoff_attempts'] > 5]

scaler = MinMaxScaler()
df.iloc[:,5:8] = scaler.fit_transform(df.iloc[:,5:8].to_numpy())

df['min_max_sum'] = df['nba_champ_per_attempt']  + df['playoff_score'] + df['playoff_score_per_attempt'] 
df.sort_values(['min_max_sum'], ascending=False, inplace=True)

df.to_csv("greatest_playoff_coach_filtered_and_min_max.csv", index=False, encoding="utf-8-sig")
df
```
